```python
a = [1] # var name a binds with list type object
b = a  # Now another variable B binds with object of var name a so both binded with object [1]
print(a) # [1]
print(b) # [1]
print(id(a)) 
print(id(b)) # Same memory id for both
print(a == b) #TRUE
print( a is b) # TRUE
```

```python
x = [1] # var name x binds with list type object 1 
y = x # another var name y binds with object of var name x so both binded with object [1]
y.append(2) # here append function is called it tells to go at var name y obejct reference point at add value 2 in the list since list is mutuable also and x and y are identical and  values are equal both can see change
print(x) # [1,2]
print(y) # [1,2]
print(id(x)) # same
print(id(y)) # same
print(x == y) # yes value changes can be seen here
print(x is y) # yes both are identical

'''Better wording  y reffers to the list obeject, and append() modifies that object , Since X refers to the same object, X also observers the modification'''
```
